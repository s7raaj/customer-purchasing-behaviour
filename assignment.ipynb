# Import Libraries and Load the Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np
df = pd.read_csv('supermarket_sales.csv') 
print(df.columns)
```

# Calculate Average Transaction Value and Create Target Variable

```python
average_transaction_value = df['Total'].mean()
print("Average transaction value:", average_transaction_value)
df['spending_category'] = np.where(df['Total'] > average_transaction_value, 1, 0)
#"1" (high-spend) and those below as "0" (low-spend).
```

# Prepare Features and Target Variable

```python
X = df.drop(columns=['spending_category', 'Total', 'Invoice ID', 'Date', 'Time']) 
y = df['spending_category'] 
X = pd.get_dummies(X, drop_first=True)
```

# Feature Scaling

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Scale the features for important for logistic regression
```

# Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

# Logistic Regression Model

```python
log_reg = LogisticRegression(max_iter=1000) # increasing iterations of logistic regression
log_reg.fit(X_train, y_train)
y_pred_log_reg = log_reg.predict(X_test)
print("Logistic Regression Classification Report:")
print(classification_report(y_test, y_pred_log_reg))
```

# Logistic Regression Confusion Matrix and Metrics

```python
confusion = confusion_matrix(y_test, y_pred_log_reg) # Confusion matrix with custom metrics for Logistic Regression
TN, FP, FN, TP = confusion.ravel()

accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
precision_log_reg = TP / (TP + FP) if (TP + FP) > 0 else 0
recall_log_reg = TP / (TP + FN) if (TP + FN) > 0 else 0
f1_score_log_reg = 2 * (precision_log_reg * recall_log_reg) / (precision_log_reg + recall_log_reg) if (precision_log_reg + recall_log_reg) > 0 else 0

print("Logistic Regression - Accuracy:", accuracy_log_reg)
print("Precision:", precision_log_reg)
print("Recall:", recall_log_reg)
print("F1 Score:", f1_score_log_reg)
```

# Decision Tree Model

```python
tree_classifier = DecisionTreeClassifier(random_state=42) # training the Descision Tree Model classifier
tree_classifier.fit(X_train, y_train)
y_pred_tree = tree_classifier.predict(X_test)
accuracy_tree = accuracy_score(y_test, y_pred_tree)
print("Decision Tree - Accuracy:", accuracy_tree)
```

# Feature Importance from Decision Tree

```python
importances = pd.DataFrame({
    'Feature': X.columns,
    'Importance': tree_classifier.feature_importances_
}).sort_values(by='Importance', ascending=False)

print("\nFeature Importance from Decision Tree:")
print(importances)
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importances)
plt.title('Feature Importance from Decision Tree')
plt.show()
# Extracting the feature importance from the Decision Tree and plotting the feature importance 
```

# Correlation Analysis

```python
numeric_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(8, 6))
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Features')
plt.show()
```

# Cross-Validation for Logistic Regression

```python
log_reg_cv = LogisticRegression(max_iter=1000)
log_reg_scores = cross_val_score(log_reg_cv, X_scaled, y, cv=5, scoring='accuracy')
print("Logistic Regression Cross-Validation Accuracy Scores:", log_reg_scores)
print("Mean Accuracy:", np.mean(log_reg_scores))
# cross validating the logistic regression
```

# Cross-Validation for Decision Tree

```python
tree_cv = DecisionTreeClassifier(random_state=42)
tree_scores = cross_val_score(tree_cv, X, y, cv=5, scoring='accuracy')
print("Decision Tree Cross-Validation Accuracy Scores:", tree_scores)
print("Mean Accuracy:", np.mean(tree_scores))
# cross validating the Decision Tree
```

# Prune Decision Tree and Feature Importance After Pruning

```python
tree_pruned = DecisionTreeClassifier(max_depth=4, random_state=42)
tree_pruned.fit(X_train, y_train)
importances_pruned = pd.DataFrame({
    'Feature': X.columns,
    'Importance': tree_pruned.feature_importances_
}).sort_values(by='Importance', ascending=False)
print("\nFeature Importance After Pruning:")
print(importances_pruned)
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importances_pruned)
plt.title('Feature Importance After Decision Tree Pruning')
plt.show()
#Pruning the Decision Tree(also setting max depth) and plotting the feature importance after pruning
```
